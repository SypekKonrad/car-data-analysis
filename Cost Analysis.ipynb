### Fuel Expenditure and Price per Kilometer

This cell calculates key metrics for fuel cost analysis:

- Total Expenditure: Cost of each refueling.
- Price per Kilometer: Cost to drive 1 kilometer.

Additionally, it computes:
- Total Fuel Expenditure: Sum of all refueling costs.
- Average Fuel Expenditure per Refueling: Mean cost of refueling.
- Average Price per Kilometer: Average cost to drive 1 kilometer.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('car_refueling_data.csv')

df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y')


df['Fuel Efficiency (km/L)'] = df['Kilometers Traveled'] / df['Liters']

df['Total Expenditure'] = df['Price'] * df['Liters']

df['Price per km'] = df['Total Expenditure'] / df['Kilometers Traveled']

# fuel_efficiency = df['Fuel Efficiency (km/L)'].mean().round(2)

total_expenditure = df['Total Expenditure'].sum()

average_expenditure_per_refueling = df['Total Expenditure'].mean()

average_price_per_km = df['Price per km'].mean()
```


### Visualizing Fuel Price Trends Over Time

- A line plot is created to depict the trend of fuel prices (in PLN per liter) over time.
- Data points are marked with circles to highlight individual measurements.
- This visualization helps identify patterns or fluctuations in fuel prices, aiding in cost analysis.

```python
plt.style.use('seaborn-v0_8-darkgrid')

plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Price'], marker='o', label='Fuel Price (per Liter)', color='b')
plt.title("Fuel Price Over Time", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Fuel Price (PLN per Liter)", fontsize=14)
plt.legend(fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Analyzing Price per Kilometer and Fuel Efficiency

- A line plot visualizes the price per kilometer (PLN) over time, with data points marked for individual refueling events.
- The average price per kilometer is highlighted with a red dashed line, making it easy to compare fluctuations against the average value.

  ## Key metrics:
        - Total Fuel Expenditure: Total cost of fuel across all refueling events.
        - Average Fuel Expenditure per Refueling: Average cost per refueling.
        - Average Price per Kilometer: Average cost of traveling one kilometer in PLN.

These insights provide a comprehensive view of travel cost efficiency.

```python
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Price per km'], marker='o', label='Price per km', color='purple')
plt.axhline(y=average_price_per_km, color='r', linestyle='--', 
            label=f'Average Price per km: {average_price_per_km:.2f} PLN')
plt.title("Price per Kilometer Over Time", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Price per Kilometer (PLN)", fontsize=14)
plt.legend(fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# print(f"Fuel Efficiency (km/L): {fuel_efficiency}")
print(f"Total Fuel Expenditure: {total_expenditure:.2f} PLN")
print(f"Average Fuel Expenditure per Refueling: {average_expenditure_per_refueling:.2f} PLN")
print(f"Average Price per Kilometer: {average_price_per_km:.2f} PLN")
```
